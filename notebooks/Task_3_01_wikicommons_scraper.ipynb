```python
# Pixabay API Scraper for Fruits & Vegetables

import os
import requests
from tqdm import tqdm

API_KEY = "54287906-cef36533ce55948f0474da79d"  # <-- Put your Pixabay API key here
BASE_DIR = r"C:\coding5final\coding5\data\images\fruit&veg"

# Categories and folder paths
categories = {
    "apples": os.path.join(BASE_DIR, "fruit", "apples"),
    "bananas": os.path.join(BASE_DIR, "fruit", "bananas"),
    "carrots": os.path.join(BASE_DIR, "veg", "carrots"),
    "tomatoes": os.path.join(BASE_DIR, "veg", "tomatoes")
}

IMAGES_PER_CATEGORY = 50
for cat_path in categories.values():
    os.makedirs(cat_path, exist_ok=True)
def download_images(keyword, save_dir, api_key, per_page=50):
    print(f"\n[!] Scraping images for '{keyword}'...")
    url = "https://pixabay.com/api/"
    params = {
        "key": api_key,
        "q": keyword,
        "image_type": "photo",
        "per_page": per_page,
        "safesearch": True
    }
    response = requests.get(url, params=params)
    
    if response.status_code != 200:
        print(f"[✗] Request failed: {response.status_code}")
        return
    
    data = response.json()
    hits = data.get("hits", [])
    
    if not hits:
        print(f"[!] No images found for '{keyword}'")
        return
    
    for hit in tqdm(hits, desc=f"Downloading {keyword}"):
        image_url = hit.get("largeImageURL")
        if not image_url:
            continue
        
        try:
            image_data = requests.get(image_url).content
            filename = os.path.join(save_dir, os.path.basename(image_url))
            with open(filename, "wb") as f:
                f.write(image_data)
        except Exception as e:
            print(f"[✗] Failed to download {image_url}: {e}")

for cat, path in categories.items():
    download_images(cat, path, API_KEY, per_page=IMAGES_PER_CATEGORY)

print("\n Completed scraping all categories!")

```
