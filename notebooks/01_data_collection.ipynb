```python
##Food Shop data collecting!

This notebook defines the category schema and collects a raw labelled dataset
of grocery item names and their corresponding categories. The aim is to
mimic the automatic categorisation behaviour of the Apple Reminders "Groceries" list.

At this stage, the data is collected in a raw form without preprocessing.
```

```python
##Category Schema

The following categories are used in this project:

- Fruit and Veg
- Eggs and Dairy
- Bakery
- Pantry
- Snacks
- Household
- Meat and Fish
- Drinks
- Frozen Foods


These categories were chosen because they are:
- Common in supermarket organisation
- Small in number 
- Similar to those used by consumer grocery applications (https://www.sciencedirect.com/science/article/abs/pii/S0195666320305481)
```

```python
import pandas as pd

data = [
    # Fruit & Vegetables
    ("apples", "Fruit & Vegetables"),
    ("bananas", "Fruit & Vegetables"),
    ("oranges", "Fruit & Vegetables"),
    ("spinach", "Fruit & Vegetables"),
    ("carrots", "Fruit & Vegetables"),
    ("potatoes", "Fruit & Vegetables"),

    # Eggs & Dairy
    ("eggs", "Eggs & Dairy"),
    ("milk", "Eggs & Dairy"),
    ("cheddar cheese", "Eggs & Dairy"),
    ("butter", "Eggs & Dairy"),
    ("yoghurt", "Eggs & Dairy"),

    # Meat & Fish
    ("chicken breast", "Meat & Fish"),
    ("beef mince", "Meat & Fish"),
    ("salmon fillet", "Meat & Fish"),
    ("pork chops", "Meat & Fish"),

    # Bakery
    ("bread", "Bakery"),
    ("bagels", "Bakery"),
    ("croissants", "Bakery"),

    # Pantry / Dry Goods
    ("pasta", "Pantry / Dry Goods"),
    ("rice", "Pantry / Dry Goods"),
    ("tinned tomatoes", "Pantry / Dry Goods"),
    ("olive oil", "Pantry / Dry Goods"),

    # Snacks
    ("crisps", "Snacks"),
    ("chocolate bar", "Snacks"),
    ("biscuits", "Snacks"),

    # Drinks
    ("orange juice", "Drinks"),
    ("coffee", "Drinks"),
    ("tea bags", "Drinks"),

    # Frozen Foods
    ("frozen peas", "Frozen Foods"),
    ("ice cream", "Frozen Foods"),

    # Household
    ("washing up liquid", "Household"),
    ("toilet paper", "Household")
]

df = pd.DataFrame(data, columns=["item", "category"])
df
```
